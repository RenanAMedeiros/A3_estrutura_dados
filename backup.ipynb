```python
from datetime import datetime, timedelta
import math

def calc_hours_diff(future_date):
    current_datetime = datetime.now()
    hours_diff = (future_date - current_datetime).total_seconds() / 3600
    return math.ceil(hours_diff)

# Fila
encomendas = []

encomendas.append({
    'id_encomenda': 'E011',
    'id_destino': 'D1',
    'prazo_entrega': calc_hours_diff(datetime.now() + timedelta(days=3)),
    'peso': 5000
})
encomendas.append({
    'id_encomenda': 'E012',
    'id_destino': 'D1',
    'prazo_entrega': calc_hours_diff(datetime.now() + timedelta(days=5)),
    'peso': 10000
})
encomendas.append({
    'id_encomenda': 'E021',
    'id_destino': 'D2',
    'prazo_entrega': calc_hours_diff(datetime.now() + timedelta(days=2)),
    'peso': 8000
})
encomendas.append({
    'id_encomenda': 'E031',
    'id_destino': 'D3',
    'prazo_entrega': calc_hours_diff(datetime.now() + timedelta(days=4)),
    'peso': 3000
})

# Reordena a fila por prazo de entrega
encomendas.sort(key=lambda x: x["prazo_entrega"])

raw_encomendas = encomendas.copy()
```
